Plot of n vs. P(n)

```python
import numpy as np
import scipy.special as scs
import matplotlib.pyplot as plt

N = 50 #number of flips
n = np.arange(0, N) #number of heads

def prob(x):
    return scs.factorial(50)/((2**N)*(scs.factorial(x)*scs.factorial(N-x)))

plt.plot(n,prob(n))
plt.xlabel("n")
plt.ylabel("P(n)")
plt.figure()
```

Plot of n vs. ln(P(n))

```python
import numpy as np
import scipy.special as scs
import matplotlib.pyplot as plt

N = 50 #number of flips
n = np.arange(0, N) #number of heads

def prob(x):
    return scs.factorial(50)/((2**N)*(scs.factorial(x)*scs.factorial(N-x)))

plt.plot(n, np.log(prob(n)))
plt.xlabel("n")
plt.ylabel("ln (P(n))")
plt.figure()
```
